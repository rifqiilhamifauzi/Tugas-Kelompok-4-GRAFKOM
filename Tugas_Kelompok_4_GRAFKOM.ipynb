```python
import matplotlib.pyplot as plt

def window_to_viewport(xw, yw, window, viewport):
    xw_min, yw_min, xw_max, yw_max = window
    xv_min, yv_min, xv_max, yv_max = viewport

    # Menghitung faktor skala
    sx = (xv_max - xv_min) / (xw_max - xw_min)
    sy = (yv_max - yv_min) / (yw_max - yw_min)

    # Pemetaan koordinat
    xv = xv_min + (xw - xw_min) * sx
    yv = yv_min + (yw - yw_min) * sy

    return xv, yv


world_x = [20, 40, 30, 20]
world_y = [20, 20, 40, 20]

# 2. Definisi Boundary
window = [10, 10, 50, 50]
viewport = [100, 100, 300, 500]

# 3. Transformasi Objek
view_x, view_y = [], []
for x, y in zip(world_x, world_y):
    vx, vy = window_to_viewport(x, y, window, viewport)
    view_x.append(vx)
    view_y.append(vy)

# 4. Visualisasi
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot Window (World)
ax1.set_title("Window (World Coordinates)")
ax1.plot(world_x, world_y, 'ro-')
ax1.set_xlim(0, 60); ax1.set_ylim(0, 60)
ax1.grid(True)

# Plot Viewport (Screen)
ax2.set_title("Viewport (Screen Coordinates)")
ax2.plot(view_x, view_y, 'bo-')
ax2.set_xlim(0, 400); ax2.set_ylim(0, 600)
ax2.grid(True)

plt.show()
```
